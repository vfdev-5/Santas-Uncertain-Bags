```python
from collections import defaultdict
```

```python
gifts = defaultdict(list)
for gift_index, gift_type in enumerate(GIFT_TYPES):
    for i in range(N_BAGS):
        gifts[gift_type].append(weight_fn(gift_index, count=1))
```

```python
plt.figure(figsize=(14,6))
for i, t in enumerate(GIFT_TYPES):
    plt.subplot(2,5,i+1)
    plt.title('%s' % t)
    plt.hist(gifts[t])
```
